```python
NumPy arrays can be created in several ways. Here are some of the most common methods:

1.array(): Creates NumPy arrays from Python lists
2.zeros(): Creates NumPy arrays with all elements initialized to 0
3.ones(): Creates NumPy arrays with all elements initialized to 1
4.full(): Creates an array with all elements initialized to same user-specified value
5.arange(): Creates NumPy arrays with a sequence of values (similar to python range() function)
6.linspace(): Creates NumPy arrays with a sequence of evenly spaced values
7.diag(): Creates NumPy array with the provided list of numbers as the diagonal elements and zeros elsewhere
8.eye(): Creates NumPy array with ones on the diagonal and zeros elsewhere
9.identity(): Creates an identity matrix
10.random.rand(): Creates NumPy array of random numbers sampled from a uniform distribution
```

# Accessing an Array


### Indexing and Slicing NumPy Arrays

NumPy arrays can be indexed and sliced in several ways depending on the number of dimensions.

### Indexing:
Indexing follows a simple rule that we may have encountered before. Let's give it a try.

1D Array: array_name[idx] (Similar to Python list)

2D Array: array_name[row_idx, col_idx] (Similar to Matrix)

ND Array: array_name[dim1_idx, dim2_idx, ..., dimN_idx]


OR

2D Array: array_name[row_idx][col_idx]

ND Array: array_name[dim1_idx][dim2_idx] ... [dimN_idx]

### Slicing:

Slicing follows the same rule as indexing. The only difference is that for each dimension we provide it with a range of indeces and sometimes (optional) step size (similar to arange() function discussed above).

1. 1D Array: array_name[idx1:idx2:step]

2. 2D Array: array_name[row_idx1:row_idx2:step1, col_idx1:col_idx2:step2]
3. ND Array: array_name[dim1_idx1:dim1_idx2:step1, ..., dimN_idx1:dimN_idx2:stepN]

```python
import numpy as np
a = np.array([1,2,3,4,5,6]) #1Dim
```
