## 2. Matrix Operations

```python
b = np.random.randint(0,10, (3,4))
```

```python
a,b
```

```python
# a @ b
res = np.dot(a,b)
```

```python
res
```

```python
# a.T
np.transpose(a)
```

```python
np.trace(b) # Diagonal sum
```

```python
# Inverse of an matrix

sq = np.random.randint(0,10, (3,3))
a_det = np.linalg.det(sq)
```

```python
a_det
```

```python
inv_sq = np.linalg.inv(sq)
```

```python
np.dot(sq, inv_sq).round()
```

### Solve Equation : 4x + 2y = 8 , 5x + 3y = 9

```python
a_inv = np.linalg.inv(a)
```
