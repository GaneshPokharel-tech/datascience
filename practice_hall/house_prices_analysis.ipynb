```python
def convert_amount(value):
    if "Lac" in value:
        return float(value.replace("Lac","") ) * 100000
    elif "Cr" in value:
        return float (value.replace("Cr","")) * 10000000
    else: 
        return np.nan
df["total_price"] = df["Amount(in rupees)"].apply(convert_amount)
```

```python
df[["Amount(in rupees)","total_price"]].head(10)
```

```python
df["total_price"].describe()
```

```python
Q1 = df["total_price"].quantile(0.25)
Q3 = df["total_price"].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3 + 1.5*IQR
lower_bound , upper_bound
```

```python
df["total_price_capped"] = np.where(
    df["total_price"]> upper_bound,
        upper_bound,
    df["total_price"]
)
```

```python
df["total_price_capped"].describe()
```

**House prices show strong right skewness due to a small number of ultra-expensive properties. Using IQR-based capping, extreme upper outliers were controlled while preserving the central distribution, resulting in a more stable and realistic price distribution for further analysis.**

```python
def convert_area(value):
    if pd.isna(value):
        return np.nan
    value = value.lower().strip()
    if "sqft" in value:
        return float(value.replace("sqft",""))
    elif "sqm" in value:
        return float(value.replace("sqm","")) *10.7639
    elif "sqyrd" in value:
        return float(value.replace("sqyrd","")) * 9
    else:
        return np.nan
df["carpet_area_sqft"]= df["Carpet Area"].apply(convert_area)
```

```python
df["Carpet Area"].dropna().sample(10)
```

```python
df.loc[
    df["Carpet Area"].str.contains("sqyrd|sqm|sqrt",case= False, na= False),
    ["Carpet Area","carpet_area_sqft"]
].sample(10)
```

```python
df["carpet_area_sqft"].isna().sum()
```

```python
df["price_per_sqft"]= df["total_price_capped"]/df["carpet_area_sqft"]
```

```python
df[["total_price_capped","carpet_area_sqft","price_per_sqft"]].sample(10)
```

Price per square foot could only be computed for properties with valid carpet area values. Observations with missing area were retained but excluded from price density analysis.

```python
df["price_per_sqft"].describe()
```

```python
df["price_per_sqft"].notna().sum()
```

Price per square foot shows a highly right-skewed distribution. A small number of extremely high values significantly inflate the mean. Hence, median and IQR-based methods are preferred for further analysis.

```python
Q1 = df["price_per_sqft"].quantile(0.25)
Q3 = df["price_per_sqft"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

Q1, Q3, IQR, lower_bound, upper_bound
```

```python
df["price_per_sqft_capped"] = df["price_per_sqft"].clip(
    lower=lower_bound,
    upper=upper_bound
)
```

```python
df["price_per_sqft_capped"].describe()
```

Price per square foot initially exhibited extreme skewness due to unrealistic values. IQR-based capping was applied to stabilize the distribution. After capping, the feature shows a reasonable range and variance, making it suitable for further analysis and modeling.

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor

num_features = ['carpet_area_sqft', 'Bathroom_clean']
cat_features = ['location_clean']

num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))
])

cat_pipeline = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_features),
        ('cat', cat_pipeline, cat_features)
    ]
)

model = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('rf', RandomForestRegressor(
        n_estimators=100,
        random_state=42,
        n_jobs=-1
    ))
])
```

```python
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

r2, rmse
```
