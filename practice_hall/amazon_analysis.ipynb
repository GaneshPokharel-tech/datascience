```python
#convert to numeric value
```

```python
amazon_df["discounted_price"] = (
    amazon_df["discounted_price"].astype(str).str.replace(",","",regex= False).str.extract(r"(-?\d+\.?\d*)") [0].astype(float)
)
```

```python
amazon_df['discounted_price'].dtype
```

```python
amazon_df["actual_price"] = (
    amazon_df["actual_price"].astype(str).str.replace(",","",regex = True).str.extract(r"(-?\d+\.?\d*)") [0].astype(float)
)
```

```python
amazon_df["actual_price"].dtypes
```

```python
amazon_df["discount_percentage"] = (
    amazon_df["discount_percentage"]
    .astype(str)                 # üîë force string
    .str.replace("%", "", regex=False)
    .astype(float)
)
```

```python
amazon_df["rating"]= pd.to_numeric(amazon_df["rating"], errors="coerce")
```

```python
amazon_df["rating_count"] = pd.to_numeric(amazon_df["rating_count"],errors= "coerce")
```
