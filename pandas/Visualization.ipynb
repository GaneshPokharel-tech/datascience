#  1. Bar Chart

```python
titanic_df ["Survived"].value_counts().plot(
    kind = "bar",
    title = "Number of Survived people",
    figsize = (4,3),
    legend = True,
    xlabel = "Survived (0=No), (1=Yes)",
    ylabel = "No. of Passangers",
    grid = True,
    yticks = list(range(0,500,50)),
    colormap = "Paired"
)
```

```python
sns.countplot(
    data = titanic_df,
    x = "Survived",
    hue = "Sex"
)
```

```python
titanic_df[titanic_df["Sex"] == "male"]["Survived"].value_counts().plot(
    kind = "bar",
    title = "Count of Male Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "Paired"
)
```

```python
group_by_gender = titanic_df[["Sex","Survived"]].groupby(by = "Sex")

group_by_gender["Survived"].value_counts().plot(
    kind = "bar",
    title = "Count of Male Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "Paired"
)
```

```python
pd.crosstab(
    titanic_df.Survived,
    titanic_df.Sex
).plot(
    kind = "bar",
    title = "Count of Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "tab20"
)
```

# 2. Pie Chart

```python
group_by_embarked = titanic_df.groupby(by = "Embarked")
group_by_embarked.count()
```

```python
group_by_embarked.count().plot(
    kind = "pie", y = "Survived",
    title = "Count of Survived Passengers",
    figsize = (4,3),
    legend = True,
    colormap = "tab20",
    autopct = '%1.1f%%',
    startangle = 90
)
```

```python
pd.crosstab(
    titanic_df.Survived,
    titanic_df.Sex
).plot(
    kind = "pie", subplots = True,
    title = "Count of Survived Passengers",
    figsize = (5,4),
    legend = True,
    colormap = "tab20"
)
```
