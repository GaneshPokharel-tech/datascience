## Handling Null or NaN or NA values

```python
titanic_df.isna().sum()
```

* Removing Null values
* Fill something ( Imputation )
* Null as new feature ( Feature Engineering )

```python
# Removing
df_filter_1 = titanic_df.dropna(subset="Embarked", axis=0, ignore_index = True)
```

```python
df_filter_1
```

```python
df_filter_1.isna().sum()
```

```python
# Null value  as feature

df_filter_1['Cabin'] = df_filter_1.Cabin.notna().astype(int)
```

```python
df_filter_1.isna().sum()
```

```python
# Fill Null values (Imputation)
print(titanic_df.Age.mean())
titanic_df.Age.plot(kind = "hist", bins = 20)
```

```python
mean_age= round(df_filter_1.Age.mean(),2)
mean_age
```

```python
# BackFill and ForwardFill and custom value
df_filter_2 = df_filter_1.copy()
df_filter_2.Age = df_filter_1["Age"].fillna(value = mean_age,inplace=False)
```

```python
df_filter_2.isna().sum()
```

```python
print(df_filter_2.Age.mean())
df_filter_2.Age.plot(kind = "hist", bins = 20)
```

### Imputation of Age wrt Gender

```python
group_object= df_filter_1.groupby(by = "Sex")
mean_age= group_object["Age"].transform("mean")
df_filter_3 = df_filter_1.copy()
df_filter_3.fillna(value={"Age":mean_age},inplace= True)
```

```python
df_filter_3.isna().sum()
```

```python
print(df_filter_3.Age.mean())
df_filter_3.Age.plot(kind = "hist", bins = 20)
```

### Handling Null value for Age, using initials (Mr., Miss or Master)groups.

```python
mean_age = df_filter_1.groupby(by = "initials")["Age"].transform("mean")

df_filter_4 = df_filter_1.copy()

df_filter_4.fillna(value= {"Age": mean_age}, inplace = True)

print(df_filter_4.Age.mean())
df_filter_4.Age.plot(kind = "hist", bins = 20)
```

```python
def get_titles(names):
    first_name = names.split(",")[-1]
    title = first_name.split(".")[0]
    return title.strip().lower()

df_filter_1["Titles"] = df_filter_1.Name.apply(get_titles)
```

```python
df_filter_1.Titles.replace(
    to_replace=["dr", "rev", "col", "don", "sir","major", "capt"] , value="mr", inplace=True
)

df_filter_1.Titles.replace(
    to_replace={"mlle":"miss", "mme": "mrs", "ms":"miss", "jonkheer": "master", "lady":"mrs","the countess":"miss"}, inplace=True
)

mean_ages = df_filter_1.groupby(by = "Titles")["Age"].transform("mean")
```

```python
df_filter_5 = df_filter_1.copy()

df_filter_5.fillna(value= {"Age": round(mean_ages,2)}, inplace = True)

print(df_filter_5.Age.mean())
df_filter_5.Age.plot(kind = "hist", bins = 20)
```
