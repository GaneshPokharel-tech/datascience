## Inerting and Deleting Column and Rows

```python
# Dropping a column
# dataframe.drop(<column_name>, axis)
# axis : {0 or 'index', 1 or 'columns'}, default 0
#     Whether to drop labels from the index (0 or 'index') or columns (1 or 'columns')

titanic_df.drop("PassengerId", axis = 1, inplace = True)
# titanic_df.drop(["PassengerId","Name"], axis = 1, inplace = True)
```

```python
titanic_df.head()
```

```python
titanic_df.drop([0,2,6],axis=0).head()
```

## Inserting or replacing a column

```python
initials= [name.split(",")[-1].split(".")[0] for name in titanic_df["Name"]]
titanic_df["initials"] = initials
titanic_df.head()
```

```python
titanic_df.insert(loc=1,column="new_initials",value= initials)
titanic_df.head()
```

### Inserting or replacing rows

```python
titanic_df.loc[887,:] = [0,"mr",3,"Ram","male",22,4,1,"T",1,"CA","S","mr"]

titanic_df.tail()
```

## Handling Null or NaN or NA values

```python
titanic_df.isna().sum()
```

* Removing Null values
* Fill something ( Imputation )
* Null as new feature ( Feature Engineering )

```python
# Removing
df_filter_1 = titanic_df.dropna(subset="Embarked", axis=0, ignore_index = True)
```

```python
df_filter_1
```

```python
df_filter_1.isna().sum()
```

```python
# Null value  as feature

df_filter_1['Cabin'] = df_filter_1.Cabin.notna().astype(int)
```

```python
df_filter_1.isna().sum()
```

```python
# Fill Null values (Imputation)
print(titanic_df.Age.mean())
titanic_df.Age.plot(kind = "hist", bins = 20)
```
