```python
address = "kathmandu"
```

```python
address[0]
```

```python
address[4]
```

```python
len(address)
```

##String Formatting:
You can use string formatting to insert variables or expressions into a string. There are several ways to format strings in Python, including using % operator, .format() and f-strings. Here are some examples:

```python
name = "ganesh"
age = 26
sent = "Your name is %s and your age is %d" %(name.capitalize(),age)
print(sent)
```

```python
sent = "your name is {} and your age is {}".format(name.capitalize(), age)
sent
```

```python
sent = f"Your name is {name} and you are {age}  Years old."
print(sent)
```

```python
color = "red"
animal = "sparrow"

prompt = f"Picture of {color} {animal}"
print(prompt)
```
