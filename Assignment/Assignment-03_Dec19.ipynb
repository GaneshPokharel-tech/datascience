### Math Tutor Using Random, Arithmetic Operators, and OOP

*Scenario:*
Create a `MathTutor` class that generates random math questions using `random` and `math operators`.

* Track correct and incorrect answers.
* Provide a score at the end.
* Handle invalid input using exception handling.

> Concepts: Random, Math, Exception Handling, Classes

```python
# Your Solution Here
import random 
class MathTutor:
    def __init__(self): 
        self.correct_count = 0
        self.incorrect_count = 0
    def play (self):
        print(f"welcome to the mathtutor game.")
        operators = ["+", "-" , "*" , "//"]
        for i in range(10):
            num1 = random.randint(1,100)
            num2 = random.randint(1,100)
            op = random.choice(operators)
            if op == "//"  and num2 == 0:
                op = random.randint(1,100)
            if op == "+":
                Answer = num1 + num2
            elif op == "-":
                Answer = num1 - num2
            elif op == "*":
                Answer = num1 * num2
            elif op == "//":
                Answer = num1 // num2
                
            try:
                print(f" Question {i+1} : {num1} {op} {num2} = ? ")
                user_input = input("Enter Your Answer: ")
                user_answer = int(user_input)

                if user_answer == Answer:
                    print("Correct Answer")
                    self.correct_count +=1
                else:
                    print(f"OOPS!!! Incorrect Answer. The Correct Answer was {Answer}.")
                    self.incorrect_count += 1

            except ValueError:
                print("Invalid Input.Enter Number only understand ")
                self.incorrect_count += 1
        self.show_score()
    
    def show_score(self):
        total = self.correct_count + self.incorrect_count
        print(f"Final Results:")
        print(f"Correct: {self.correct_count}")
        print(f"Incorrect: {self.incorrect_count}")


tutor = MathTutor()
tutor.play()

            
        
```

### Exception Handling Scenario: Online Age-Restricted Service

**Scenario:**
You’re building a sign-up system for an online movie rental platform. Some movies are age-restricted (18+). You need to ensure proper validation and error handling during user registration.

**Task:**

1. Create a custom exception class called `UnderageError` that inherits from `Exception`.
2. Write a function `register_user()` that:

   * Takes a user’s `name` and `age` as input.
   * Raises `UnderageError` if the user is under 18.
   * Otherwise, prints a welcome message.
3. Wrap the function call in a `try` block and handle the exception.
4. Use `else` to confirm successful registration and `finally` to always print **“Thank you for using MovieTime!”** regardless of outcome.

Also try to validate if the age input is numeric. Raise a `ValueError` if not, and handle it separately.



```python
# Your Solution Here
class UnderAgeError(Exception):
    """Exception raised when a user is under 18."""
    pass

def register_user(name, age_input):
    if not str(age_input).isdigit():  
        raise ValueError("Age must be Numeric Value.")
    age = int(age_input)
    if age < 18:
        raise UnderAgeError("under 18 age not allowed to registration.")
    else: 
        print(f"welcome, {name} You have successfully registered.")

try:
    user_name = input("enter your name : ")
    user_age = input("enter your age : ")
    register_user(user_name, user_age) 
except ValueError as ve:
    print("validation error age must be in numeric value.")
except UnderAgeError as uae:
    print("access denied")
else:
    print("registration confirmed. Enjoy your movies")
finally:
    print("Thank you for using MovieTime!")
```

# Problem Statement

We want to build an online shopping cart system that allows users to add products to their cart, calculate the total cost, apply discounts, and generate an invoice. The system should include the following functionalities:

- Adding products to the cart
- Removing products from the cart
- Calculating the total cost
- Applying discounts based on user type
- Generating an invoice

### 1. Create the Product class

We create a basic `Product` class with attributes for the product name and price.

```python
# Your Solution Here
class product:
    def __init__(self,name,price):
        self.name = name
        self.price = price
```

### 2.  Implement the User class

In this step, we create a `User` class with attributes for the user's name and whether they are a premium member.

```python
# Your Solution Here
class user:
    def __init__(self,name,premium):
        self.name = name
        self.premium = premium
```

### 3. Create the ShoppingCart class

In this step, we create a `ShoppingCart` class with methods for adding and removing products from the cart, as well as calculating the total cost of the items in the cart.

**Note**: Define `calculate_total_cost` method in the `ShoppingCart` class, that applies a `10%` discount to the total cost if you are `premium` User.

```python
# Your Solution Here
class ShoppingCart:
    def __init__(self,user):
        self.user = user
        self.products = []
    def add_product(self,product):
        self.products.append(product)
    def remove_product(self,product):
        if product in self.products:
            self.products.remove(product)
    def calculate_total_cost(self):
        total = sum(product.price for product in self.products)
        if self.user.premium:
            return total*0.9
        else:
            return total 
```

### 4. Testing the functionality

Now that we have implemented the necessary classes and methods, let's test our online shopping cart system:

```python
# Your Solution Here
'''creat user profile'''
regular_user = user("Bishnu", False)
premium_user = user("Ganesh", True)

'''creat product profile'''
product1 = product("wine", 1000)
product2 = product("8848 full", 2000)
product3 = product("rice", 3000)

'''creat shoping cart '''
cart_regular = ShoppingCart(regular_user)
cart_premium = ShoppingCart(premium_user)

'''add product to  regular cart'''
cart_regular.add_product(product1)
cart_regular.add_product(product2)
cart_regular.add_product(product3)
print("total for regular user: ", cart_regular.calculate_total_cost())

'''add product to premium cart '''
cart_premium.add_product(product1)
cart_premium.add_product(product2)
cart_premium.add_product(product3)
print("total for premium user : ", cart_premium.calculate_total_cost())

'''remove product from regular cart'''
cart_regular.remove_product(product1)
print("total amount after remove for regular user: ", cart_regular.calculate_total_cost())


```

### 5. Generating Invoice for a given cart

```python
# Your Solution Here

def generate_invoice(cart):
    print(f"invoice for {cart.user.name}")
    print("products: ")
    for product in cart.products:
        print(f" {product.name} : {product.price} ")
    print(f"total cost: {cart.calculate_total_cost()} ")
    if cart.user.premium:
        print("10% discount applied")
generate_invoice(cart_regular)

generate_invoice(cart_premium)
```
