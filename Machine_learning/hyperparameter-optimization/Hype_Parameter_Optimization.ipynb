```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


scaler = StandardScaler()
log_res = LogisticRegression()

data = pd.read_csv("diabetes.csv")
X = data.drop("Outcome", axis = 1).to_numpy()
Y = data[["Outcome"]].to_numpy()

scaled_X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    scaled_X, Y, test_size=0.2, random_state=989, stratify=Y
)
```

# Hyperparameters
Hyperparameters are parameters whose values control the learning process and determine the values of model parameters that a learning algorithm ends up learning. The prefix ‘hyper_’ suggests that they are ‘top-level’ parameters that control the learning process and the model parameters that result from it.

### Hyper Parameter Testing
1.GridSearch

2.RandomSearch

```python
from sklearn.model_selection import ParameterGrid, GridSearchCV, RandomizedSearchCV

hyperparameter = {
    "class_weight" : ["balanced", None],
    "penalty" : ["l1","l2"],
    "max_iter" : [50,100,150,200,250],
    "solver" :["liblinear"]
}

len(ParameterGrid(hyperparameter))
```

```python
# Grid Search

model = LogisticRegression()
```

```python
grid_search = GridSearchCV(
    estimator= model,
    param_grid= hyperparameter,
    scoring= "recall",
    cv= 3, 
    n_jobs= 8
)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model
```

## Random Search

```python
random_search = RandomizedSearchCV(
    estimator= model,
    param_distributions= hyperparameter,
    n_iter = 15,
    scoring= "accuracy",
    cv = 3,
    n_jobs=-1
)

random_search.fit(X_train, y_train)
```

```python
random_search.best_params_
```

```python
random_search.best_score_
```

```python
random_search.best_estimator_
```
