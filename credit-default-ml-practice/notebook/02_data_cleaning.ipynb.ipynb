```python
import pandas as pd

df = pd.read_excel(
    "../data/raw/default of credit card clients.xls",
    header=1
)

df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["ID"])
```

```python
df
```

### Dropping Identifier Column
- `ID` is an identifier, not a predictive feature
- Removed to prevent noise in model training

```python
df.value_counts("EDUCATION")
```

```python
df.value_counts("MARRIAGE")
```

```python
df["EDUCATION"] = df["EDUCATION"].replace([0,5,6],4)
df["MARRIAGE"] = df["MARRIAGE"].replace(0,3)
```

### Handling Invalid Categorical Values
- Invalid EDUCATION values (0, 5, 6) merged into category 4 (others)
- Invalid MARRIAGE value (0) merged into category 3 (others)
- Rows preserved to avoid data loss


```python
df["EDUCATION"].value_counts(normalize=True)
```

```python
df["MARRIAGE"].value_counts(normalize=True)
```

```python
df.to_csv("../data/processed_credit_card_clients.csv", index=False)
```
